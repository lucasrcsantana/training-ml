## Calculating distance between two users

```python
ratings.query('userId == 1')[['movieId', 'rating']].set_index('movieId').head()
```

```python
def get_user_ratings(user_id: int) -> pd.DataFrame:
    user_ratings = ratings.query(f'userId == {user_id}')[['movieId', 'rating']].set_index('movieId')
    return user_ratings
```

```python
first_user = get_user_ratings(1)
second_user = get_user_ratings(4)
```

```python
compare_ratings = first_user.join(second_user, lsuffix='_first', rsuffix='_second', how='inner')
```

```python
compare_ratings.head()
```

```python
distance = np.linalg.norm(compare_ratings['rating_first'] - compare_ratings['rating_second'])
```

```python
def distance_two_users(first_user_id: int, second_user_id: int, min_ocurrencies = 5) -> float:
    first_user = get_user_ratings(first_user_id)
    second_user = get_user_ratings(second_user_id)
    compare_ratings = first_user.join(second_user, lsuffix='_first', rsuffix='_second', how='inner')
    
    if len(compare_ratings) < min_ocurrencies:
        distance = None
    else:
        distance = np.linalg.norm(compare_ratings['rating_first'] - compare_ratings['rating_second'])
    
    return [first_user_id, second_user_id, distance]
```

```python
distance_two_users(1, 2)
```

```python
def distance_one_to_many_users(first_user_id: int, n_users: int=None) -> pd.DataFrame:
    if n_users:
        users_id = ratings['userId'].unique()[:n_users]
    else:
        users_id = ratings['userId'].unique()
    distances = [distance_two_users(first_user_id, user_id) for user_id in users_id]
    distances = pd.DataFrame(distances, columns = ['first_user_id', 'second_user_id', 'distance'])
    distances = distances.dropna()
    return distances
```

```python
def nearest_users(first_user_id: int, n_users: int=None) -> pd.DataFrame:
    distances = distance_one_to_many_users(first_user_id, n_users)
    distances = distances.set_index('second_user_id').drop(first_user_id, errors='ignore').sort_values('distance')
    return distances
```

```python
df_nearest_users = nearest_users(52, 22)
df_nearest_users
```

```python
def user_recommendation(first_user_id: int, n_users: int=None) -> pd.DataFrame:
    df_nearest_users = nearest_users(first_user_id, n_users)
    nearest_user = df_nearest_users.iloc[0].name
    movies_nearest_user = get_user_ratings(nearest_user)
    movies_first_user = get_user_ratings(first_user_id)
    movies_recommended = movies_nearest_user.drop(movies_first_user.index, errors='ignore').sort_values('rating', ascending = False)
    movies_recommended = movies_recommended.join(movies)
    return movies_recommended
```
