```python
votes = ratings['movieId'].value_counts()
```

```python
movies['total_votes'] = votes
```

```python
movies.head()
```

```python
movies.sort_values('total_votes', ascending=False)
```

```python
mean_ratings = ratings.groupby('movieId').mean()['rating']
```

```python
movies['mean_ratings'] = mean_ratings
```

```python
movies.head()
```

```python
movies.sort_values('mean_ratings', ascending=False)
```
